### 2. Scrape the details team India’s international fixtures from bcci.tv.
Url = https://www.bcci.tv/.
You need to find following details:
A) Match title (I.e. 1st ODI)
B) Series
C) Place
D) Date
E) Time
Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
webD = webdriver.Chrome(r"C:\Users\User\Downloads\chromedriver_win32\chromedriver.exe")
url = "https://www.bcci.tv/"
webD.get(url)
time.sleep(2)

search_btn = webD.find_element_by_class_name('navigation__item  ')
search_btn.click()
time.sleep(2)

webD.find_element_by_class_name('drop-down__option  ').click()
time.sleep(2)

for _ in range(500):
    webD.execute_script("window.scrollBy(0,10000)")
    
dicty = {}
dicty['MatchTitle'] = []
dicty['Series'] = []
dicty['Place'] = []
dicty['Date'] = []
dicty['Time'] = []

elements1 = webD.find_elements_by_class_name('fixture__additional-info')

for i in elements1:
    dicty['Series'].append(i.text.split('\n')[0])
    dicty['Place'].append(i.text.split('\n')[1])
    
elements2 = webD.find_elements_by_class_name('fixture__full-date')

for i in elements2:
    j = i.text.split('\n')[0] + " " + i.text.split('\n')[1]
    dicty['Date'].append(j)
    dicty['Time'].append(i.text.split('\n')[2])
    
elements3 = webD.find_elements_by_class_name('fixture__teams')

for i in elements3:
    dicty['MatchTitle'].append(i.text.replace('\n', ' '))
    
bcci_df = pd.DataFrame({'Match Title': dicty['MatchTitle'], 'Series': dicty['Series'], 
                        'Place': dicty['Place'], 'Date': dicty['Date'], 'Time': dicty['Time']})
```

### 3. Scrape the details of selenium exception from guru99.com.
Url = https://www.guru99.com/
You need to find following details:
A) Name
B) Description
Note: - From guru99 home page you have to reach to selenium exception handling page through code.

```python
webD = webdriver.Chrome(r"C:\Users\User\Downloads\chromedriver_win32\chromedriver.exe")
url = "https://www.guru99.com/"
webD.get(url)
time.sleep(2)

elements = webD.find_elements_by_xpath('//li[@class = "fa fa-chevron-circle-right"]//a')
time.sleep(2)

urls = []
for i in elements:
    urls.append(i.get_attribute("href"))
    
dicty = {}
dicty['head'] = []
dicty['desc'] = []

for i in urls:
    
    webD.get(i)
    try:
        h = webD.find_element_by_class_name('page-header')
        dicty['head'].append(h.text)
    except NoSuchElementException:
        dicty['head'].append('-')
        
    try:
        elements = webD.find_element_by_class_name('item-page')
        string = elements.text.split("What should I know?")[0]
        try:
            string = string.split("Summary")[1]
        except:
            pass
        string.replace('\n','')            
        dicty['desc'].append(string)
    except NoSuchElementException:
        dicty['desc'].append('-')
```

```python
guru99_df = pd.DataFrame({'Name': dicty['head'], 'Description': dicty['desc']})
```

```python
guru99_df
```

### 5. Scrape the details of trending repositories on Github.com.
Url = https://github.com/
You have to find the following details:
A) Repository title
B) Repository description
C) Contributors count
D) Language used
Note: - From the home page you have to click on the trending option from Explore menu through code.

```python
webD = webdriver.Chrome(r"C:\Users\User\Downloads\chromedriver_win32\chromedriver.exe")
url = "https://github.com/"
webD.get(url)
time.sleep(2)

elements = webD.find_elements_by_xpath('//summary[@class = "HeaderMenu-summary HeaderMenu-link px-0 py-3 border-0 no-wrap d-block d-lg-inline-block"]')
elements[1].click()
time.sleep(2)

elements1 = webD.find_elements_by_xpath('//li[@class = "edge-item-fix"]//a')

for i in elements1:
    link = i.get_attribute('href')
    if '/trending' in link:
        i.click()
        break
        
dicty = {}
dicty['Repositorytitle'] = []
dicty['Repositorydescription'] = []
dicty['Contributorscount'] = []
dicty['Language'] = []

elements2 = webD.find_elements_by_xpath('//h1[@class = "h3 lh-condensed"]//a')

for i in elements2:
    dicty['Repositorytitle'].append(i.text)
    
elements3 = webD.find_elements_by_xpath('//p[@class = "col-9 color-text-secondary my-1 pr-4"]')

for i in elements3:
    dicty['Repositorydescription'].append(i.text)

elements4 = webD.find_elements_by_xpath('//div[@class = "f6 color-text-secondary mt-2"]')

for i in elements4:
    dicty['Language'].append(i.text.split(" ")[0])
    dicty['Contributorscount'].append(i.text.split(" ")[2])
```

```python
github_df = pd.DataFrame({'Repository title': dicty['Repositorytitle'], 'Repository description': dicty['Repositorydescription'], 
                          'Contributors count': dicty['Contributorscount'], 'Language': dicty['Language'],})
github_df
```
